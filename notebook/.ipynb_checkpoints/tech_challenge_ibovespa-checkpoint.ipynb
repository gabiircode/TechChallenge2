# ðŸ”¹ Etapa 1 â€” Processamento dos Dados

```python
#Principais bibliotecas utilizadas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
# Leitura e tratamento
dados = pd.read_csv("historic_data.csv")
dados["Data"] = pd.to_datetime(dados["Data"], dayfirst=True)
dados = dados.sort_values("Data").reset_index(drop=True)

for col in ['Ãšltimo', 'Abertura', 'MÃ¡xima', 'MÃ­nima']: #-----colunas de preÃ§o
    dados[col] = dados[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    dados[col] = pd.to_numeric(dados[col], errors='coerce')

dados["Var%"] = dados["Var%"].astype(str).str.replace('%', '').str.replace(',', '.', regex=False).astype(float) #ConversÃ£o da variaÃ§Ã£o percentual

def converter_volume(v):
    if isinstance(v, str):
        v = v.replace('.', '').replace(',', '.')
        if 'M' in v: return float(v.replace('M', '')) * 1_000_000
        elif 'B' in v: return float(v.replace('B', '')) * 1_000_000_000
    return pd.to_numeric(v, errors='coerce')

dados["Vol"] = dados["Vol."].apply(converter_volume)
```

## ðŸ”¹ Etapa 2 - EDA

```python
# GrÃ¡fico de linha mostrando a tendÃªncia histÃ³rica
plt.figure(figsize=(12, 4))
plt.plot(dados['Data'], dados['Ãšltimo'], color='midnightblue')
plt.title("HistÃ³rico Temporal do IBOVESPA")
plt.xlabel("Data")
plt.ylabel("Fechamento")
plt.grid(True)
plt.tight_layout()
plt.show()

# Engenharia para variÃ¡veis de correlaÃ§Ã£o
for i in range(1, 6):
    dados[f'lag_{i}'] = dados['Ãšltimo'].shift(i)

dados['sma_5'] = dados['Ãšltimo'].rolling(window=5).mean()

delta = dados['Ãšltimo'].diff()
gain = delta.clip(lower=0).rolling(window=14).mean()
loss = -delta.clip(upper=0).rolling(window=14).mean()
rs = gain / loss
dados['rsi'] = 100 - (100 / (1 + rs))

# Mapa de calor com variÃ¡veis tÃ©cnicas
plt.figure(figsize=(10, 6))
sns.heatmap(
    dados[['Ãšltimo', 'sma_5', 'rsi', 'lag_1', 'lag_2']].dropna().corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("CorrelaÃ§Ã£o entre VariÃ¡veis TÃ©cnicas")
plt.tight_layout()
plt.show()
```

# ðŸ”¹ Etapa 3 â€” Engenharia de Features e DefiniÃ§Ã£o do Target

```python
# CriaÃ§Ã£o de variÃ¡veis defasadas (lags) do preÃ§o de fechamento
# Cada 'lag' representa o valor do fechamento de dias anteriores (atÃ© 5 dias atrÃ¡s)
for i in range(1, 6):
    dados[f'lag_{i}'] = dados['Ãšltimo'].shift(i)

# CÃ¡lculo de mÃ©dias mÃ³veis simples (SMA)
# A SMA ajuda a suavizar flutuaÃ§Ãµes diÃ¡rias e identificar tendÃªncias
dados['sma_5'] = dados['Ãšltimo'].rolling(window=5).mean()   # mÃ©dia dos Ãºltimos 5 dias
dados['sma_10'] = dados['Ãšltimo'].rolling(window=10).mean() # mÃ©dia dos Ãºltimos 10 dias

# CÃ¡lculo do RSI (Ãndice de ForÃ§a Relativa)
# O RSI Ã© um indicador tÃ©cnico que mede a velocidade e a mudanÃ§a dos movimentos de preÃ§o
delta = dados['Ãšltimo'].diff()                              # diferenÃ§a entre dias consecutivos
gain = delta.clip(lower=0).rolling(window=14).mean()        # ganhos mÃ©dios dos Ãºltimos 14 dias
loss = -delta.clip(upper=0).rolling(window=14).mean()       # perdas mÃ©dias dos Ãºltimos 14 dias
rs = gain / loss                                            # razÃ£o entre ganho e perda
dados['rsi'] = 100 - (100 / (1 + rs))                       # normalizaÃ§Ã£o do RSI entre 0 e 100

# CÃ¡lculo do MACD (Moving Average Convergence Divergence)
# O MACD ajuda a identificar mudanÃ§as na forÃ§a, direÃ§Ã£o e duraÃ§Ã£o de uma tendÃªncia
ema_12 = dados['Ãšltimo'].ewm(span=12, adjust=False).mean()  # mÃ©dia mÃ³vel exponencial de 12 dias
ema_26 = dados['Ãšltimo'].ewm(span=26, adjust=False).mean()  # mÃ©dia mÃ³vel exponencial de 26 dias
dados['macd'] = ema_12 - ema_26                             # linha MACD
dados['macd_signal'] = dados['macd'].ewm(span=9, adjust=False).mean()  # linha de sinal

# CriaÃ§Ã£o da variÃ¡vel de fechamento e da variaÃ§Ã£o percentual do dia seguinte
dados['Fechamento Hoje'] = dados['Ãšltimo']
dados['pct_change'] = dados['Fechamento Hoje'].pct_change().shift(-1)

# DefiniÃ§Ã£o da variÃ¡vel alvo (target)
# Se a variaÃ§Ã£o do dia seguinte for maior que 0,7%, consideramos uma alta relevante (target = 1)
dados['target'] = (dados['pct_change'] > 0.007).astype(int)

# Remove linhas com valores ausentes resultantes dos cÃ¡lculos anteriores
dados = dados.dropna()
```

# ðŸ”¹ Etapa 4 â€” Treinamento e Teste (Ãºltimos 30 dias como teste)

```python
# Seleciona apenas as colunas que serÃ£o usadas como variÃ¡veis preditoras (features)
# Remove colunas informativas que nÃ£o devem entrar no modelo como 'Data', 'Ãšltimo', 'target' etc.
features = [col for col in dados.columns if col not in [
    'Data', 'Ãšltimo', 'Abertura', 'MÃ¡xima', 'MÃ­nima',
    'Vol.', 'Var%', 'Fechamento Hoje', 'pct_change', 'target'
]]

# Define X com as features e y com o valor que queremos prever (alta ou baixa)
X = dados[features]
y = dados['target']

# Divide os dados em treino e teste com base no tempo (os Ãºltimos 30 dias serÃ£o usados como teste)
X_train = X.iloc[:-30]
y_train = y.iloc[:-30]
X_test = X.iloc[-30:]
y_test = y.iloc[-30:]

# Aplica padronizaÃ§Ã£o nas variÃ¡veis (transforma tudo para a mesma escala)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Seleciona automaticamente as variÃ¡veis mais importantes com base no modelo XGBoost
selector = SelectFromModel(XGBClassifier(random_state=42), threshold='median')
selector.fit(X_train_scaled, y_train)
X_train_sel = selector.transform(X_train_scaled)
X_test_sel = selector.transform(X_test_scaled)

# Calcula o peso da classe minoritÃ¡ria para tratar desbalanceamento da base (mais zeros do que uns)
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

# Define os hiperparÃ¢metros que serÃ£o testados no modelo (grade de busca)
param_grid = {
    'max_depth': [3, 5, 7],              # profundidade da Ã¡rvore
    'n_estimators': [100, 200],          # nÃºmero de Ã¡rvores (estimadores)
    'learning_rate': [0.01, 0.1],        # taxa de aprendizado (quanto cada Ã¡rvore aprende)
    'subsample': [0.8, 1.0]              # fraÃ§Ã£o de dados usada por Ã¡rvore (controle de overfitting)
}

# Cria divisÃµes temporais para validaÃ§Ã£o cruzada, respeitando a ordem do tempo
tscv = TimeSeriesSplit(n_splits=5)

# Define o modelo base com os parÃ¢metros fixos e o peso calculado
model = XGBClassifier(
    objective='binary:logistic',         # tipo de problema: classificaÃ§Ã£o binÃ¡ria
    eval_metric='logloss',              # mÃ©trica de avaliaÃ§Ã£o
    use_label_encoder=False,            # evita aviso desnecessÃ¡rio
    scale_pos_weight=scale_pos_weight,  # ajuda a equilibrar as classes
    random_state=42
)

# Busca pelos melhores hiperparÃ¢metros usando validaÃ§Ã£o cruzada com base temporal
grid_search = GridSearchCV(
    model, param_grid, cv=tscv,
    n_jobs=-1, verbose=1, scoring='accuracy'
)
grid_search.fit(X_train_sel, y_train)

# Seleciona o melhor modelo encontrado na busca
best_model = grid_search.best_estimator_

# Avalia o desempenho do modelo nos dados de treino
acc_train = accuracy_score(y_train, best_model.predict(X_train_sel))

# Avalia o desempenho do modelo nos Ãºltimos 30 dias (teste real)
acc_test = accuracy_score(y_test, best_model.predict(X_test_sel))

# Realiza validaÃ§Ã£o cruzada para garantir estabilidade do modelo
cv_scores = cross_val_score(best_model, X_train_sel, y_train, cv=tscv, scoring='accuracy')
```

# ðŸ”¹ Etapa 5 â€” Resultado Final e PrevisÃ£o

```python

from datetime import timedelta

ultimo_dia = dados.iloc[-1]
fechamento_hoje = float(ultimo_dia['Fechamento Hoje'])

X_final = selector.transform(scaler.transform(X.tail(1)))
predito = best_model.predict(X_final)[0]

previsao_pct = float(ultimo_dia['pct_change'])
previsao_valor = fechamento_hoje * (1 + previsao_pct)
tendencia_prevista = "Alta ðŸ“ˆ" if predito == 1 else "Baixa ðŸ“‰"

resultado = {
    "Data da PrevisÃ£o": (ultimo_dia['Data'] + timedelta(days=1)).strftime("%Y-%m-%d"),
    "Fechamento Hoje": f"R$ {fechamento_hoje:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
    "PrevisÃ£o de Fechamento AmanhÃ£": f"R$ {previsao_valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
    "TendÃªncia Prevista": tendencia_prevista,
    "AcurÃ¡cia Treino": f"{acc_train:.2%}",
    "AcurÃ¡cia Teste (30 dias)": f"{acc_test:.2%}",
    "AcurÃ¡cia CV MÃ©dia": f"{cv_scores.mean():.2%}",
    "Desvio CV": f"{cv_scores.std():.2%}"
}

for chave, valor in resultado.items():
    print(f"{chave}: {valor}")
```
