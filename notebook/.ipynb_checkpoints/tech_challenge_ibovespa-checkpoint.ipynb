# 🔹 Etapa 1 — Processamento dos Dados

```python
#Principais bibliotecas utilizadas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
# Leitura e tratamento
dados = pd.read_csv("historic_data.csv")
dados["Data"] = pd.to_datetime(dados["Data"], dayfirst=True)
dados = dados.sort_values("Data").reset_index(drop=True)

for col in ['Último', 'Abertura', 'Máxima', 'Mínima']: #-----colunas de preço
    dados[col] = dados[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    dados[col] = pd.to_numeric(dados[col], errors='coerce')

dados["Var%"] = dados["Var%"].astype(str).str.replace('%', '').str.replace(',', '.', regex=False).astype(float) #Conversão da variação percentual

def converter_volume(v):
    if isinstance(v, str):
        v = v.replace('.', '').replace(',', '.')
        if 'M' in v: return float(v.replace('M', '')) * 1_000_000
        elif 'B' in v: return float(v.replace('B', '')) * 1_000_000_000
    return pd.to_numeric(v, errors='coerce')

dados["Vol"] = dados["Vol."].apply(converter_volume)
```

## 🔹 Etapa 2 - EDA

```python
# Gráfico de linha mostrando a tendência histórica
plt.figure(figsize=(12, 4))
plt.plot(dados['Data'], dados['Último'], color='midnightblue')
plt.title("Histórico Temporal do IBOVESPA")
plt.xlabel("Data")
plt.ylabel("Fechamento")
plt.grid(True)
plt.tight_layout()
plt.show()

# Engenharia para variáveis de correlação
for i in range(1, 6):
    dados[f'lag_{i}'] = dados['Último'].shift(i)

dados['sma_5'] = dados['Último'].rolling(window=5).mean()

delta = dados['Último'].diff()
gain = delta.clip(lower=0).rolling(window=14).mean()
loss = -delta.clip(upper=0).rolling(window=14).mean()
rs = gain / loss
dados['rsi'] = 100 - (100 / (1 + rs))

# Mapa de calor com variáveis técnicas
plt.figure(figsize=(10, 6))
sns.heatmap(
    dados[['Último', 'sma_5', 'rsi', 'lag_1', 'lag_2']].dropna().corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlação entre Variáveis Técnicas")
plt.tight_layout()
plt.show()
```

# 🔹 Etapa 3 — Engenharia de Features e Definição do Target

```python
# Criação de variáveis defasadas (lags) do preço de fechamento
# Cada 'lag' representa o valor do fechamento de dias anteriores (até 5 dias atrás)
for i in range(1, 6):
    dados[f'lag_{i}'] = dados['Último'].shift(i)

# Cálculo de médias móveis simples (SMA)
# A SMA ajuda a suavizar flutuações diárias e identificar tendências
dados['sma_5'] = dados['Último'].rolling(window=5).mean()   # média dos últimos 5 dias
dados['sma_10'] = dados['Último'].rolling(window=10).mean() # média dos últimos 10 dias

# Cálculo do RSI (Índice de Força Relativa)
# O RSI é um indicador técnico que mede a velocidade e a mudança dos movimentos de preço
delta = dados['Último'].diff()                              # diferença entre dias consecutivos
gain = delta.clip(lower=0).rolling(window=14).mean()        # ganhos médios dos últimos 14 dias
loss = -delta.clip(upper=0).rolling(window=14).mean()       # perdas médias dos últimos 14 dias
rs = gain / loss                                            # razão entre ganho e perda
dados['rsi'] = 100 - (100 / (1 + rs))                       # normalização do RSI entre 0 e 100

# Cálculo do MACD (Moving Average Convergence Divergence)
# O MACD ajuda a identificar mudanças na força, direção e duração de uma tendência
ema_12 = dados['Último'].ewm(span=12, adjust=False).mean()  # média móvel exponencial de 12 dias
ema_26 = dados['Último'].ewm(span=26, adjust=False).mean()  # média móvel exponencial de 26 dias
dados['macd'] = ema_12 - ema_26                             # linha MACD
dados['macd_signal'] = dados['macd'].ewm(span=9, adjust=False).mean()  # linha de sinal

# Criação da variável de fechamento e da variação percentual do dia seguinte
dados['Fechamento Hoje'] = dados['Último']
dados['pct_change'] = dados['Fechamento Hoje'].pct_change().shift(-1)

# Definição da variável alvo (target)
# Se a variação do dia seguinte for maior que 0,7%, consideramos uma alta relevante (target = 1)
dados['target'] = (dados['pct_change'] > 0.007).astype(int)

# Remove linhas com valores ausentes resultantes dos cálculos anteriores
dados = dados.dropna()
```

# 🔹 Etapa 4 — Treinamento e Teste (últimos 30 dias como teste)

```python
# Seleciona apenas as colunas que serão usadas como variáveis preditoras (features)
# Remove colunas informativas que não devem entrar no modelo como 'Data', 'Último', 'target' etc.
features = [col for col in dados.columns if col not in [
    'Data', 'Último', 'Abertura', 'Máxima', 'Mínima',
    'Vol.', 'Var%', 'Fechamento Hoje', 'pct_change', 'target'
]]

# Define X com as features e y com o valor que queremos prever (alta ou baixa)
X = dados[features]
y = dados['target']

# Divide os dados em treino e teste com base no tempo (os últimos 30 dias serão usados como teste)
X_train = X.iloc[:-30]
y_train = y.iloc[:-30]
X_test = X.iloc[-30:]
y_test = y.iloc[-30:]

# Aplica padronização nas variáveis (transforma tudo para a mesma escala)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Seleciona automaticamente as variáveis mais importantes com base no modelo XGBoost
selector = SelectFromModel(XGBClassifier(random_state=42), threshold='median')
selector.fit(X_train_scaled, y_train)
X_train_sel = selector.transform(X_train_scaled)
X_test_sel = selector.transform(X_test_scaled)

# Calcula o peso da classe minoritária para tratar desbalanceamento da base (mais zeros do que uns)
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

# Define os hiperparâmetros que serão testados no modelo (grade de busca)
param_grid = {
    'max_depth': [3, 5, 7],              # profundidade da árvore
    'n_estimators': [100, 200],          # número de árvores (estimadores)
    'learning_rate': [0.01, 0.1],        # taxa de aprendizado (quanto cada árvore aprende)
    'subsample': [0.8, 1.0]              # fração de dados usada por árvore (controle de overfitting)
}

# Cria divisões temporais para validação cruzada, respeitando a ordem do tempo
tscv = TimeSeriesSplit(n_splits=5)

# Define o modelo base com os parâmetros fixos e o peso calculado
model = XGBClassifier(
    objective='binary:logistic',         # tipo de problema: classificação binária
    eval_metric='logloss',              # métrica de avaliação
    use_label_encoder=False,            # evita aviso desnecessário
    scale_pos_weight=scale_pos_weight,  # ajuda a equilibrar as classes
    random_state=42
)

# Busca pelos melhores hiperparâmetros usando validação cruzada com base temporal
grid_search = GridSearchCV(
    model, param_grid, cv=tscv,
    n_jobs=-1, verbose=1, scoring='accuracy'
)
grid_search.fit(X_train_sel, y_train)

# Seleciona o melhor modelo encontrado na busca
best_model = grid_search.best_estimator_

# Avalia o desempenho do modelo nos dados de treino
acc_train = accuracy_score(y_train, best_model.predict(X_train_sel))

# Avalia o desempenho do modelo nos últimos 30 dias (teste real)
acc_test = accuracy_score(y_test, best_model.predict(X_test_sel))

# Realiza validação cruzada para garantir estabilidade do modelo
cv_scores = cross_val_score(best_model, X_train_sel, y_train, cv=tscv, scoring='accuracy')
```

# 🔹 Etapa 5 — Resultado Final e Previsão

```python

from datetime import timedelta

ultimo_dia = dados.iloc[-1]
fechamento_hoje = float(ultimo_dia['Fechamento Hoje'])

X_final = selector.transform(scaler.transform(X.tail(1)))
predito = best_model.predict(X_final)[0]

previsao_pct = float(ultimo_dia['pct_change'])
previsao_valor = fechamento_hoje * (1 + previsao_pct)
tendencia_prevista = "Alta 📈" if predito == 1 else "Baixa 📉"

resultado = {
    "Data da Previsão": (ultimo_dia['Data'] + timedelta(days=1)).strftime("%Y-%m-%d"),
    "Fechamento Hoje": f"R$ {fechamento_hoje:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
    "Previsão de Fechamento Amanhã": f"R$ {previsao_valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
    "Tendência Prevista": tendencia_prevista,
    "Acurácia Treino": f"{acc_train:.2%}",
    "Acurácia Teste (30 dias)": f"{acc_test:.2%}",
    "Acurácia CV Média": f"{cv_scores.mean():.2%}",
    "Desvio CV": f"{cv_scores.std():.2%}"
}

for chave, valor in resultado.items():
    print(f"{chave}: {valor}")
```
